```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
%matplotlib inline
```

```python
x = np.linspace(-10, 10)
y1 = (x**5 - 5*x**4 -30*x**3 + 110*x**2 + 29*x - 105)

y2 = (5*x**4 - 20*x**3 -90*x**2 + 220*x + 29)

y3 = ((x-2)*x*(x**4-4*x**3-53*x**2+114*x+315))/6

plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.xlim(-10,10)
plt.ylim(-2750,2750)

plt.grid()
plt.title('Aufgabe1')
plt.legend(['f(x) = x^5-5x^4-30x^3+110x^2+29x-105', 'f\'(x) = 5x^4-20x^3-90x^2+220x+29', 'F(x) = ((x-2)*x*(x^4-4x^3-53x^2+114x+315))/6'])
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
