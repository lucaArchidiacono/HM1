```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
%matplotlib inline
```

```python
def polynomfunc(a, xmin, xmax):
    shapes = np.shape(a)
    x = np.linspace(xmin, xmax)
    p = 0
    dp = 0
    pint = 0
    column = len(shapes) - 1
 
    if shapes[0] <= 0 or (column > 1 and shapes[column] > 1):
        raise Exception('Fehler')
    for index, item in enumerate(a):
        if type(item) is int or float:
            shape = shapes[0] - 1
            p += (item*(x**(shape - index)))
            dp += ((item*(shape - index)*(x**((shape-index)-1))))
            pint += ((item / ((shape-index)+1))*(x**((shape-index)+1)))
        else:
            shape = shapes[0] - 1
            p += (item[0]*(x**(shape - index)))
            dp += ((item[0]*(shape - index)*(x**((shape-index)-1))))
            pint += ((item[0] / ((shape-index)+1))*(x**((shape-index)+1)))
    return (x, p, dp, pint)
```

```python
a =  np.array([1,-5,-30,110,29,-105])
xmin = -10
xmax = 10

[x,p,dp,pint] = polynomfunc(a,xmin,xmax)

plt.figure()
plt.plot(x,p)
plt.plot(x,dp)
plt.plot(x,pint)

plt.xlim(xmin, xmax)
plt.ylim(-2750,2750)

plt.grid()
plt.title('Aufgabe2')
plt.legend(['f(x) = x^5-5x^4-30x^3+110x^2+29x-105', 'f\'(x) = 5x^4-20x^3-90x^2+220x+29', 'F(x) = ((x-2)*x*(x^4-4x^3-53x^2+114x+315))/6'])
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
