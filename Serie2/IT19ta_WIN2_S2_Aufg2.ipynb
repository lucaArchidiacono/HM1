```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
x = np.linspace(1.99, 2.01, 501)
y1 = (x**7 - 14*x**6 +84*x**5-280*x**4 + 560*x**3 - 672*x**2 + 448*x - 128)
y2 = (x-2)**7

plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.grid()
plt.title('Aufgabe 2a')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
Python berechnet die Funktionen verschieden.
Möglicherweise sind die Kommawerte zu lang. Diese werte werden somit nach jedem Rechnungsschritt gerundet. Dies führt zu einer relativ hohen differenz zwischen den beiden Polynomen.
```

```python

steps = round((10**-14 - -10**-14)/10**-17)
x = np.linspace(-10**-14, 10**-14, steps)
y1 = x/(np.sin(1+x) - np.sin(1))

plt.figure()
plt.plot(x, y1)

plt.grid()
plt.title('Aufgabe 2b')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
Nein. Bei der Berechnung des Grenzwertes ist uns aufgefallen, dass bei x=0 ein instabiles Resultat erfolgt.
```

```python
steps = round((10**-14 - -10**-14)/10**-17)
x = np.linspace(-10**-14, 10**-14, steps)
y1 = x/(2*np.cos(((1+x)+(1))/2)*np.sin(((1+x)-(1))/2))

plt.figure()
plt.plot(x, y1)

plt.grid()
plt.title('Aufgabe 2c')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
Der Graph sieht schoener aus als vorher. Es hat weniger unterbrueche (ausser natuerlich bei x=0). Der Grenzwert Richtung 0 ist (1/cos(1)).
```
