```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
def fooA(r):
    oneSide = np.empty(r)
    allSides = np.empty_like(oneSide)
    x = np.empty_like(oneSide)
    oneSide[0]=1
    allSides[0]=6
    x[0] = 6
    for i in range(1,r):
        oneSide[i] = np.sqrt(2-2*np.sqrt(1-(oneSide[i-1])**2/4))
        x[i] = 2**i*6
        allSides[i] = oneSide[i]*x[i]
        
    return [x, allSides]
    

x,y = fooA(30)

plt.figure()
plt.plot(x, y)
plt.grid()
plt.title('Aufgabe 3a')
plt.show()
```

```python
def fooB(r):
    oneSide = np.empty(r)
    allSides = np.empty_like(oneSide)
    x = np.empty_like(oneSide)
    oneSide[0]=1
    allSides[0]=6
    x[0] = 6
    for i in range(1,r):
        oneSide[i] = np.sqrt((oneSide[i-1]**2)/(2*(1+np.sqrt(1-oneSide[i-1]**2/4))))
        x[i] = 2**i*6
        allSides[i] = oneSide[i]*x[i]
        
    return [x, allSides]
    

x,y = fooB(30)

plt.figure()
plt.plot(x, y)
plt.grid()
plt.title('Aufgabe 3b')
plt.show()
```
