```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
def fooA(r):
    oneSide = [1]
    allSides = [6]
    x = [6]
    for i in range(1,r):
        oneSide.append(np.sqrt(2-2*np.sqrt(1-(oneSide[i-1])**2/4)))
        x.append(2**i*6)
        allSides.append(oneSide[i]*x[i])
        
    return [x, allSides]
    
x,y = fooA(30)

plt.figure()
plt.plot(x, y)
plt.grid()
plt.title('Aufgabe 3a')
plt.show()
```

```python
def fooB(r):
    oneSide = [1]
    allSides = [6]
    x = [6]
    for i in range(1,r):
        oneSide.append(np.sqrt((oneSide[i-1]**2)/(2*(1+np.sqrt(1-oneSide[i-1]**2/4)))))
        x.append(2**i*6)
        allSides.append(oneSide[i]*x[i])
        
    return [x, allSides]
    

x1,y2 = fooB(30)
print(x1, y2)

plt.figure()
plt.plot(x, y)
plt.plot(x1,y2)
plt.grid()
plt.title('Aufgabe 3b')
plt.show()
```
