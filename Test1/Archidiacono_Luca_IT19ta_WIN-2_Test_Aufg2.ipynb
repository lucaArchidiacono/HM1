```python
import numpy as np
import matplotlib.pyplot as plt

#Abbildung 1
x = np.arange(-5, 5, 0.05)
y = np.exp(x)

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Abbildung 2
x1 = np.arange(-10, 10, 0.1)
y1 = (x1**5)+(3*(x1**4))+(3*(x1**2))+x1+1

plt.plot(x1, y1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Abbildung 3
x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = 0.5*np.sin(3*x)
y1 = 0.5*np.cos(3*x)

plt.plot(x, y)
plt.plot(x, y1)
plt.show()
```
